| Name | Type | Source | Description |
| :--- | :--- | :--- | :--- |
| **adf** | file | adf.parquet | <br>This file contains data on slaughtered cattle that presented suspicious tuberculosis lesions, including the contact route through the farms where they stayed during their lifetime. <br><br> |
| vig_id | variable | adf/vig_id | <br>Unique identifier for the lesion collection procedure. <br><br> |
| vig_date | variable | adf/vig_date | <br>Collection date of the lesion sample.<br><br> |
| result | variable | adf/result | <br>qPCR (quantitative Polymerase Chain Reaction) confirmation result for bovine tuberculosis (Mycobacterium bovis).<br><br> |
| result_date | variable | adf/result_date | <br>Release date of the laboratory diagnostic result<br><br> |
| bovine_id | variable | adf/bovine_id | <br>Unique identifier of the bovine sampled (also know as sentinel) <br><br> |
| farm_id | variable | adf/farm_id | <br> Unique identifier of the contact farm <br><br> |
| contact_order | variable | adf/contact_order | <br>Sequential order of contact between the sentinel animal and the contact farm, tracked retrospectively from the slaughterhouse back to the animal's birth.<br><br> |
| last_contact | variable | adf/last_contact | <br>Date of the last contact between the animal and the contact farm.<br><br> |
| exposure_duration | variable | adf/exposure_duration | <br>Total period of exposure during which the animal remained on the contact farm.<br><br> |
| herd_size | variable | adf/herd_size | <br>Time-weighted average of the herd size on the contact farm during the exposure period.<br><br> |
| **fex** | file | fex.parquet | <br> This file contains field examination records for all animals within the contact farms. <br><br> |
| farm_id | variable | fex/farm_id | <br> Unique identifier for the farm, consistent with other datasets. <br><br> |
| field_exame_id | variable | fex/field_exame_id | <br> Unique identifier for the field examination. <br><br> |
| field_exame_date | variable | fex/field_exame_date | <br> Date when the field examination was performed. <br><br> |
| pos | variable | fex/pos | <br> Number of animals that tested positive in the field examination. <br><br> |
| nex | variable | fex/nex | <br> Number of animals examined (tested) during the procedure. <br><br> |
| Ntotal | variable | fex/Ntotal | <br> Total herd size on the farm at the date of the examination. <br><br> |
