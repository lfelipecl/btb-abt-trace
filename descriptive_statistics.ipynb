```python
#Modules
import pandas as pd
import duckdb as db
```

```python
#Loading data

mr = db.query("SELECT * FROM 'model_result.parquet'").df()
```

```python
# Table 1 - Variables description
vars_to_analyze = ['herd_size', 'exposure_duration', 'route_size', 'p_deterministic']


descriptive_table = mr[vars_to_analyze].describe().transpose()


descriptive_table.columns = ['Count', 'Mean', 'SD', 'Min', '25% (Q1)', '50% (Median)', '75% (Q3)', 'Max']


descriptive_table = descriptive_table.round(4)

descriptive_table.drop(columns=['Count'], inplace=True)

display(descriptive_table)
```
